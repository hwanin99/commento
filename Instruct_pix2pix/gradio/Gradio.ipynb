# Gradio

```python
# prompt = 'Turn it into a creature with a white belly and a red face and back.'
from instruct_pix2pix.utils.inference import generate

def gradio(img,prompt):
    return generate(checkpoint_path = "./instruct_pix2pix_model/",
           image_path = f'./image2sketch/dataset/sketch/{img}', # Change Original Image path
           prompt = prompt,
           gradio = True
          )


iface = gr.Interface(fn=gradio,
                     inputs=[gr.Textbox(label = "Image filename.jpg"),
                             gr.Textbox(label = "Prompt")
                     ],
                     outputs="image")

iface.launch()
```
